```python
from phenom_to_df import *
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
top_df = csv_to_df('test/')
glu_df = csv_to_df('bsm_test/')
```

```python
top_df = top_df.fillna(0)
top_vars = [entry for entry in top_df.columns if entry[0] == 'E'] + [entry for entry in top_df.columns if entry[0:2] == 'pt'] + [entry for entry in top_df.columns if entry[0:2] == 'et'] + [entry for entry in top_df.columns if entry[0:2] == 'ph']
```

```python
glu_df = glu_df.fillna(0)
glu_vars = [entry for entry in glu_df.columns if entry[0] == 'E'] + [entry for entry in glu_df.columns if entry[0:2] == 'pt'] + [entry for entry in glu_df.columns if entry[0:2] == 'et'] + [entry for entry in glu_df.columns if entry[0:2] == 'ph']
```

```python
top_df_E = top_df[[entry for entry in top_df.columns if entry[0] == 'E']].values.astype('float')
top_df_pt = top_df[[entry for entry in top_df.columns if entry[0:2] == 'pt']].values.astype('float')
top_df_et = top_df[[entry for entry in top_df.columns if entry[0:2] == 'et']].values.astype('float')
top_df_ph = top_df[[entry for entry in top_df.columns if entry[0:2] == 'ph']].values.astype('float')
```

```python
glu_df_E = glu_df[[entry for entry in glu_df.columns if entry[0] == 'E']].values.astype('float')
glu_df_pt = glu_df[[entry for entry in glu_df.columns if entry[0:2] == 'pt']].values.astype('float')
glu_df_et = glu_df[[entry for entry in glu_df.columns if entry[0:2] == 'et']].values.astype('float')
glu_df_ph = glu_df[[entry for entry in glu_df.columns if entry[0:2] == 'ph']].values.astype('float')
```

```python
top_E = [j for i in range(len(top_df_E)) for j in top_df_E[i]]
top_pt = [j for i in range(len(top_df_pt)) for j in top_df_pt[i]]
top_et = [j for i in range(len(top_df_et)) for j in top_df_et[i]]
top_ph = [j for i in range(len(top_df_ph)) for j in top_df_ph[i]]
```

```python
glu_E = [j for i in range(len(glu_df_E)) for j in glu_df_E[i]]
glu_pt = [j for i in range(len(glu_df_pt)) for j in glu_df_pt[i]]
glu_et = [j for i in range(len(glu_df_et)) for j in glu_df_et[i]]
glu_ph = [j for i in range(len(glu_df_ph)) for j in glu_df_ph[i]]
```

```python
tops = [top_E, top_pt, top_et, top_ph]
glus = [glu_E, glu_pt, glu_et, glu_ph]
```

```python
var_names = ['E', 'pt', 'et', 'ph']
```

```python
for i in range(len(tops)):
    sns.distplot(tops[i], color = 'y')
    sns.distplot(glus[i], color = 'black')
    plt.xlabel(var_names[i])
    plt.show()
```
